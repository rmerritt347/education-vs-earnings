```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('/Users/rasheedamerritt/Downloads/education_income.csv')

# Optional: Clean column names for consistency
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('__', '_', regex=False)

df.head()
```

```python
# Strip commas and dollar signs and convert to numeric
df['earnings_hs_grad'] = pd.to_numeric(df['earnings_hs_grad'].replace(r'[\$,]', '', regex=True), errors='coerce')
df['earnings_bachelor'] = pd.to_numeric(df['earnings_bachelor'].replace(r'[\$,]', '', regex=True), errors='coerce')
df['earnings_grad_prof'] = pd.to_numeric(df['earnings_grad_prof'].replace(r'[\$,]', '', regex=True), errors='coerce')
```

```python
import matplotlib.pyplot as plt

# Create a dictionary with average earnings by education level
education_income = {
    'High School Grad': df['earnings_hs_grad'].mean(),
    'Bachelor\'s Degree': df['earnings_bachelor'].mean(),
    'Graduate Degree': df['earnings_grad_prof'].mean()
}

# Plot
plt.bar(education_income.keys(), education_income.values(), color='skyblue')
plt.title('Average Median Earnings by Education Level')
plt.ylabel('Earnings ($)')
plt.tight_layout()
plt.show()
```
